```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from matplotlib.ticker import FuncFormatter
%config Completer.use_jedi = False
```

```python
from pathlib import Path

outputs = Path('plots/')
outputs.mkdir(exist_ok=True)
```

```python
def sigmoid(x): 
    return 1 / (1 + np.exp(-x))

with plt.xkcd():
    fig = plt.figure(dpi=200)
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines.right.set_color('none')
    ax.spines.top.set_color('none')
    x = np.linspace(0,10)
    y = sigmoid(-5+x)
    plt.plot(x,y, )
    
    
    xy = np.array([x[-15],y[-15]])
    ax.annotate(
        "PERCENTAGE OF PUTIN'S ASS \n ALREADY HANDED TO HIM",
        xy=xy, 
        xytext=xy + [4,-.3],
        font = 'Comic Sans MS',
        arrowprops=dict(arrowstyle="-|>",
                    connectionstyle="angle3,angleA=10,angleB=120")
    )
    
    x = np.linspace(0,10)
    y = sigmoid(-5+x*1.3) * .7 + .3
    plt.plot(x,y,c='red' )
    
    xy = np.array([x[-24],y[-24]])
    ax.annotate(
        "MY OVERALL SATISFACTION \n IN LIFE",
        xy=xy, 
        xytext=xy + [0,-.8],
        font = 'Comic Sans MS',
        arrowprops=dict(arrowstyle="-|>",
                    connectionstyle="angle3,angleA=-45,angleB=120")
    )
    
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_xlabel('Time')
    ax.set_xticks([])
    

    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    ax.plot((1), (0), ls="", marker=">", ms=10, color="k",
            transform=ax.get_yaxis_transform(), clip_on=False, zorder=3)
    plt.savefig(outputs/'putins-ass-vs-life-satisfaction.png',
               bbox_inches="tight")
    plt.show()
    
```
